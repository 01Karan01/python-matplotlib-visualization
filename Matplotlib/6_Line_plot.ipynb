```python
from matplotlib import pyplot as plt
import pandas as pd

data = pd.read_csv("Line_plot.csv")
data
```

```python
ages_x = data['Age']
all_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages_x, all_salaries, label='All Developers')
plt.plot(ages_x, py_salaries, label='Python')
plt.plot(ages_x, js_salaries, label='JavaScript')

plt.title('Median Salary (USD) by Age')
plt.xlabel('Age')
plt.ylabel('Salary (USD)')

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
ages_x = data['Age']
all_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages_x, all_salaries, label='All Developers')
plt.plot(ages_x, py_salaries, label='Python')
plt.plot(ages_x, js_salaries, label='JavaScript')

overall_medium = 57287

plt.fill_between(ages_x, py_salaries, overall_medium,
                where=(py_salaries>overall_medium),
                 interpolate=True, alpha=0.25)

plt.fill_between(ages_x, py_salaries, overall_medium,
                where=(py_salaries<=overall_medium),
                 interpolate=True, alpha=0.25)

plt.title('Median Salary (USD) by Age')
plt.xlabel('Age')
plt.ylabel('Salary (USD)')

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
