# Pie Charts 

```python
from matplotlib import pyplot as plt

slices=[60,40]
labels=['Sixity','Forty']

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})   #colors to the edge

plt.title('Pie chart')
plt.tight_layout()
plt.show
```

# Drawing charts (without user color)

```python
from matplotlib import pyplot as plt

slices=[60,40,25,10]
labels=['Sixity','Forty','Twenty Five','Ten']

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})   #colors to the edge

plt.title('Pie chart')
plt.tight_layout()
plt.show()
```

# Drawing charts (with user color)

```python
from matplotlib import pyplot as plt

slices=[60,40,25,10]
labels=['Sixity','Forty','Twenty Five','Ten']
colors=['#008fd5','#fc4f30','#e5ae37','#6d904f']

plt.pie(slices,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'})   #colors to the edge

plt.title('Pie chart')
plt.tight_layout()
plt.show()
```

## Explode is used to pull out the slices from the pie chart

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]    #0  slice stays in place 
                           #0.1  slice moves outward by 10% of the pies radius 
                           #Larger values  slice moves farther away.

plt.pie(slices , labels = labels , shadow = True, 
        explode = explode , wedgeprops = {'edgecolor':'black'})
plt.tight_layout()
plt.title('Pie chart for diff Programming language')
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]    #0  slice stays in place 
                           #0.1  slice moves outward by 10% of the pies radius 
                           #Larger values  slice moves farther away.

plt.pie(slices , labels = labels , shadow = True, startangle = 90,
        explode = explode , wedgeprops = {'edgecolor':'black'})
plt.tight_layout()
plt.title('Pie chart for diff Programming language')
plt.show()
```

### Arguments 1: 'Shadow' for giving the shadow to the pie chart is above
### Arguments 2: 'Startangle' for keeping the 90` one first 

## Adding % to the pie chart using "autopct='%1.1f%%'"

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0,0,0,0.1,0]    #0  slice stays in place 
                           #0.1  slice moves outward by 10% of the pies radius 
                           #Larger values  slice moves farther away.

plt.pie(slices , labels = labels , shadow = True, startangle = 90, autopct = '%1.1f%%',
        explode = explode , wedgeprops = {'edgecolor':'black'})
plt.tight_layout()
plt.title('Pie chart for diff Programming language')
plt.show()
```
