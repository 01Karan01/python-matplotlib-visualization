## Scatter Plot with real data

```python
from matplotlib import pyplot as plt
import pandas as pd
plt.style.use('seaborn-v0_8-whitegrid')

data = pd.read_csv('scatterplot.csv')
data 
```

```python
x = data['view_count']
y = data['likes']

ratio = data['ratio']
plt.scatter(x, y, c = ratio, edgecolor = 'black')
plt.xlabel('View count')
plt.ylabel('Total likes')
plt.title('Trending YT videos')
plt.show()
```

```python
x = data['view_count']
y = data['likes']

ratio = data['ratio']
plt.scatter(x, y, c = ratio, cmap = 'summer', edgecolor = 'black', linewidth = 1, alpha =0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.xlabel('View count')
plt.ylabel('Total likes')
plt.title('Trending YT videos')
plt.show()
```
