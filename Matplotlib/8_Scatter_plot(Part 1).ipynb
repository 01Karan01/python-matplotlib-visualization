## Adding color and diff shape to the point

```python
x= [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y= [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x ,y ,s = 50,c = 'green',marker = 'x')
plt.show()
```

## Setting a bar & "c" is used to Color each point based on its numeric value using the Greens colormap.
### for eg
### Higher number → darker green
### Lower number → lighter green

```python
x= [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y= [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

plt.scatter(x, y, s= 50.0,c = colors,cmap='Greens', edgecolor= 'black', linewidth= 1, alpha= 0.75)
cbar= plt.colorbar()
cbar.set_label('Satisfaction')
plt.show()
```

## Size of points depending on the sizes

```python
x= [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y= [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors= [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes= [209, 486, 381, 255, 191, 315, 185, 228, 174, 
        538, 239, 399, 153, 273, 293, 436, 501, 397, 539, 300]

plt.scatter(x, y, s= sizes,c = colors,cmap='Greens', edgecolor= 'black', linewidth= 1, alpha= 0.75)
cbar= plt.colorbar()
cbar.set_label('Satisfaction')
plt.show()
```
